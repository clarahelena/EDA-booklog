# Consolidação dos Resultados (Baseline vs KNN vs SVD)

Este notebook reúne os resultados de todos os testes efetuados pela equipe para comparar os modelos e definir qual será utilizado em Produção (P2) e reportado no Artigo (P3).

```python
import pandas as pd

# A Baseline que executamos no Notebook 02 (Cosine Similarity puro) obteve Precision@10 de 1.0000
# Na prática, ela é equivalente a um modelo KNN com K=10 usando métrica de Cosseno.
baseline_results = pd.DataFrame([{
    'algoritmo': 'Baseline (Cosseno Puro)',
    'configuracao': 'top_10',
    'precision@10': 1.0000,
    'coverage': 0.0036 # Aproximado, similar ao KNN=10
}])

# Carregando os resultados dos parceiros
df_knn = pd.read_csv('../../Machine Learning/data/processed/resultados_knn.csv')
df_knn['configuracao'] = 'k=' + df_knn['k'].astype(str) + ' | ' + df_knn['metrica']

df_svd = pd.read_csv('../../Machine Learning/data/processed/resultados_svd.csv')
df_svd['configuracao'] = 'n_comp=' + df_svd['n_components'].astype(str) + ' | var=' + df_svd['variancia_explicada'].astype(str)

# Padronizando as colunas para o Join
df_knn = df_knn[['algoritmo', 'configuracao', 'precision@10', 'coverage']]
df_svd = df_svd[['algoritmo', 'configuracao', 'precision@10', 'coverage']]

# Concatenando todos os resultados
df_final = pd.concat([baseline_results, df_knn, df_svd], ignore_index=True)

# Ordenando para descobrir o grande vencedor
df_final = df_final.sort_values(by=['precision@10', 'coverage'], ascending=[False, False])

display(df_final)
```

## Vencedor Oficial e Próximos Passos
Com base na tabela acima, observamos que o **KNN com métrica de Cosseno (ou a Baseline pura)** obteve precisão perfeita (1.0000) nos testes de aderência aos gêneros. 

O SVD, embora tenha diminuído brutalmente o tamanho do espaço dimensional (de 1100 colunas para 50), sacrificou um pouco da precisão (caiu para ~0.9903). Isso é normal no SVD (redução de ruído causa perda de um pouco de informação estrita). 

E o modelo vencedor a ser implementado no app em produção será o **KNN com k=10 e distância de Cosseno**!

```python
# Exportando a tabela final consolidada para o Artigo
df_final.to_csv('../../Machine Learning/data/processed/tabela_final_resultados.csv', index=False)
print("Tabela final exportada com sucesso!")
```
