# Notebook 0: Limpeza dos Dados

Projeto Booklog

Objetivo deste notebook:
O dataset original (data.csv) retirado do Kaggle possui 6.810 registros de livros. No entanto, esses dados brutos contêm alguns valores ausentes que podem prejudicar as análises futuras.

Este notebook foca exclusivamente na etapa de Pré-processamento. Ao final, exportaremos uma base de dados limpa para ser consumida pelos próximos notebooks de Análise Exploratória (EDA).

```python
import pandas as pd

dadosBrutos = pd.read_csv("datasets/data.csv")

dadosBrutos.info()
```

#Tratamento de Nulos

Notamos que a coluna subtitle possui mais de 4.000 valores nulos. Como ela não será vital para o sistema de recomendação inicial, vamos ignorá-la.

Porém, as colunas title, authors, categories, average_rating e dentre outras são cruciais para o projeto. Como a quantidade de nulos nessas colunas é pequena em relação a um universo de quase 7.000 livros, a estratégia mais segura para não enviesar os dados é remover as linhas que contêm esses valores vazios.

```python
colunasObrigatorias = ['title', 'authors', 'categories', 'description', 'published_year', 'average_rating', 'num_pages', 'ratings_count']
dadosLimpos = dadosBrutos.dropna(subset=colunasObrigatorias).copy()

linhasAntes = len(dadosBrutos)
linhasDepois = len(dadosLimpos)
print(f"Total de livros originais: {linhasAntes}")
print(f"Total de livros após limpeza: {linhasDepois}")
```

#Exportação do Dataset Tratado

Com os dados limpos, exportamos o dataset tratado, assim, os futuros notebooks utilizarão este novo arquivo.

```python
dadosLimpos.to_csv("datasets/books.csv", index=False)

print(f"O arquivo books foi gerado.")
```
